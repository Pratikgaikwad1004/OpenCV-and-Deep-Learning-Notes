```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import cv2
```

```python
img = cv2.imread('deadpool.jpg')
```

```python
type(img)
```

```python
plt.imshow(img)
```

```python
# Matpltlib reads image in RGB form --> Red Green Blue
# OpenCV reads image in BGR form --> Blue Green Red
```

```python
# We need to trnsform this image into RGB form
fix_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(fix_img)
```

```python
# Read grayscale image direcly
img_gray = cv2.imread('deadpool.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gray)
```

```python
img_gray.shape
```

```python
plt.imshow(img_gray, cmap="gray")
```

```python
# Resize Image

new_img = cv2.resize(fix_img, (2000, 800))
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
# Resize by Ratio

w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(fix_img, (0,0), fix_img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

```python
new_img.shape
```

```python
new_img = cv2.flip(fix_img, 0)
plt.imshow(new_img)
```
