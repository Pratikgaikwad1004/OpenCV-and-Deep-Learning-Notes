```python
import cv2
```

```python
img1 = cv2.imread("deadpool.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('deadpool name.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 15))
axes[0].imshow(img1)
axes[1].imshow(img2)
```

```python
img1 = cv2.resize(img1, (1200, 1200))
img2 = cv2.resize(img2, (1200, 1200))
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 15))
axes[0].imshow(img1)
axes[1].imshow(img2)
```

```python
blended = cv2.addWeighted(src1=img1, alpha=0.9, src2=img2, beta=0.3, gamma=0)
```

```python
plt.figure(figsize=(6, 15))
plt.imshow(blended)
```

```python
# OVERLAY SMALL IMAGE ON TOP OF LARGER IMAGE ( NO BLENDING )
```

```python
img1 = cv2.imread("deadpool.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('deadpool name.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20, 15))
axes[0].imshow(img1)
axes[1].imshow(img2)
```

```python
large_img = img1
small_img = img2
```

```python
x_offset = 0
y_offset = 0
```

```python
x_end = x_offset + small_img.shape[1]
y_end = y_offset + small_img.shape[0]
```

```python
large_img[y_offset:y_end, x_offset:x_end] = small_img
```

```python
plt.imshow(large_img)
```

```python
# BLEND TOGETHER IMAGES OF DIFFERENT SIZES
```

```python
img1 = cv2.imread("deadpool.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

img2 = cv2.imread('deadpool name.png')
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
img2.shape
```

```python
img1.shape
```

```python
x_offset = 2560 - 860
y_offset = 1600 - 397
```

```python
roi = img1[y_offset:1600, x_offset:2560]
```

```python
plt.imshow(roi)
```

```python
img2gray = cv2.cvtColor(img2, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(img2gray, cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
```

```python
plt.imshow(mask_inv, cmap='gray')
```

```python
mask_inv.shape
```

```python
import numpy as np
```

```python
white_background = np.full(img2.shape, 255, dtype=np.uint8)
```

```python
# White Background
```

```python
bk = cv2.bitwise_or(white_background, white_background, mask=mask_inv)
```

```python
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2, img2, mask=mask_inv)
```

```python
plt.imshow(fg)
```

```python
final_roi = cv2.bitwise_or(roi, fg)
```

```python
plt.imshow(final_roi)
```
